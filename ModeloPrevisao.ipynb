```python
# Obtendo dados históricos da ação PETR4.SA nos últimos 1 ano
dados = yf.Ticker("PETR4.SA").history("1y")
dados
```

```python
dados.describe()
```

## Criação do modelo de Mahcine Learning

```python
# Resetando o índice para trazer a coluna de datas como uma coluna regular
treinamento = dados.reset_index()

#  Convertendo a coluna 'Date' para o formato de data
treinamento['Date'] = treinamento['Date'].dt.date

# Selecionando apenas as colunas 'Date' e 'Close' e renomeando-as
treinamento = treinamento[['Date', 'Close']]
treinamento.columns = ['ds', 'y']

# Exibe o dataframe
treinamento
```

```python
# Inicializando o modelo Prophet
modelo = Prophet()
```

```python
treinamento
```

```python
# Treinando o modelo com os dados históricos
modelo.fit(treinamento)
```

```python
# Criando datas futuras para previsão (30 dias adicionais)
datas_futuras = modelo.make_future_dataframe(30)
datas_futuras
```

```python
# Fazendo previsões com o modelo
previsoes = modelo.predict(datas_futuras)
```

```python
# Plotando as previsões
fig = plot_plotly(modelo, previsoes)
```

```python
# Melhorando a visualização das previsões adicionando os componentes
fig2 = plot_components_plotly(modelo, previsoes)
```

```python
# Exibindo os gráficos
fig.show()
fig2.show()
```

```python
# Salvando o gráfico como arquivo HTML
pio.write_html(fig, file='/content/drive/MyDrive/UNAMA/Material_Extra/Oficina: DS e VS/previsao_petr4.html')
```
